```python
import copy
df1 = pd.read_csv(fullname)
frame1 = copy.deepcopy(df1)
frame2 = copy.deepcopy(df2)
```

```python
frame1.head(5)
l=[]
for ind in frame1.index :
    reel = float(str(frame1["sec"][ind])+"."+str(frame1["nsec"][ind]).zfill(9)) 
    rounded = round(reel,1)
    l.append(rounded)
frame1["timestamp"]=l

frame1.head(5)
```

```python
l=[]
for ind in frame2.index :
    reel = float(str(frame2["sec"][ind])+"."+str(frame2["nsec"][ind]).zfill(9)) 
    rounded = round(reel,1)
    l.append(rounded)
len(l)
frame2["timestamp"]=l
frame2.head(5)
```

```python
f=frame1.merge(frame2,on='timestamp',how = 'outer')
f.head(60)
```

```python
a = f[f["anchorId"]==0]
b = f[f["anchorId"]==1]
c = f[f["anchorId"]==2]
d = f[f["anchorId"]==3]
```

```python
at= a[["timestamp","range"]]
bt=b[["timestamp","range"]]
ct=c[["timestamp","range"]]
dt=d[["timestamp","range"]]
dt[dt["timestamp"]==12.5]
atr=at.rename(columns={"range":"anch0"},inplace=False)
btr=bt.rename(columns={"range":"anch1"},inplace=False)
ctr=ct.rename(columns={"range":"anch2"},inplace=False)
dtr=dt.rename(columns={"range":"anch3"},inplace=False)
at.head()
```

```python
final = frame1.merge(atr,on="timestamp",how='outer').merge(btr,on="timestamp",how='outer').merge(ctr,on="timestamp",how='outer').merge(dtr,on="timestamp",how='outer')
final1 = pd.concat([frame1,atr,btr,ctr,dtr],axis=1,join="outer")
data = "database.csv"
path = os.path.join(savepath,data)
final[["X","Y","timestamp","anch0","anch1","anch2","anch3"]].to_csv(path,index = False)
```
